##Stemming

```
from nltk.stem import PorterStemmer
ps = PorterStemmer()
dir(ps)
```

list of function , we are interested in this stem function

```
print(ps.stem('coder'))
print(ps.stem('coding'))
print(ps.stem('code'))
```

```
print(ps.stem('data'))
print(ps.stem('datum'))
```

```
print(ps.stem('bowl'))
print(ps.stem('bowling'))
print(ps.stem('bowler'))
```

```
def stemming(text):
  d = [ps.stem(word) for word in text]
  return d
```

```
df['review_stemmed'] = df['review_no_stopwords'].apply(lambda x: stemming(x))
df.head()
```

```
df.head(10)
```

Lemmatization is quite intensive and a slower process, but more accurate. Stemming may be more useful in queries for databases whereas Lemmatization works much better when trying to determine the text sentiment

##WordNetLemmatizer

```
import nltk
wn = nltk.WordNetLemmatizer()
ps = nltk.PorterStemmer()
dir(wn)
```

Problem with stemming that is solved by lemmatizer

```
print(ps.stem('goose'))
print(ps.stem('geese'))
```

```
print(wn.lemmatize('goose'))
print(wn.lemmatize('geese'))
```

```
print(wn.lemmatize('cactus'))
print(wn.lemmatize('cacti'))
```

```
print(ps.stem('cactus'))
print(ps.stem('cacti'))
```

```
def lemmatization(text):
  d = [wn.lemmatize(word) for word in text]
  return d
df['review_lemmatized'] = df['review_no_stopwords'].apply(lambda x : lemmatization(x))
```

Since we have list of tokens till now we need sentence with all the processing. 

```
def clean_text(txt):
  txt = "".join([c for c in txt if c not in string.punctuation])
  tokens = re.split('\W+',txt)
  txt = " ".join([wn.lemmatize(word) for word in tokens if word not in stop])
  return txt
df['sentence'] = df['review'].apply(lambda x: clean_text(x))
df.head()
```

```
cv1 = CountVectorizer(ngram_range=(2,2))

X = cv1.fit_transform(df['sentence'])
print(X.shape)
```
