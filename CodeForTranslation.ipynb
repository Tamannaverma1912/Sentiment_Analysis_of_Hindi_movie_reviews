```python
df= pd.read_csv('final_testing_data.csv')
df
```

```python
df= df.drop(columns= ['Unnamed: 0'])
df
```

```python
sentiment = {'postive': 1,'negative': 0} 
df.sentiment = [sentiment[item] for item in df.sentiment] 

df
```

```python
df_copy=df.copy()
df_copy= df_copy.drop(columns= ['sentiment'])
df_copy
```

```python
from googletrans import Translator

translator=Translator()
```

```python
translations = {}
#0-110,115-120, 125-160,165-370, 375-390, 395-475, 480-605, 615-645,650-810, 815-820, 825-853
for column in df_copy.columns:
    # Unique elements of the column
    unique_elements = df_copy[column].unique()
    for element in unique_elements:
        # Adding all the translations to a dictionary (translations)
        translations[element] = translator.translate(element).text
translations
```

```python
df.replace(translations, inplace = True)
df.head(10)
```

```python
# remove punctuation
import string
df['Text'] = df['Text'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
import re
def tokenize(txt):
  tokens = re.split('\W+',txt)
  return tokens

df['review'] = df['Text'].apply(lambda x: tokenize(x.lower()))
```

```python
df
```

```python
import nltk
nltk.download('stopwords')
#from nltk.corpus import stopwords
stop = stopwords.words('english')

def remove_stopwords(txt_tokenized):
  txt_clean = [word for word in txt_tokenized if word not in stop]
  return txt_clean

df['review'] = df['review'].apply(lambda x: remove_stopwords(x))
```

```python
import nltk

nltk.download('wordnet')
```

```python
wn = nltk.WordNetLemmatizer()
def clean_text_sentence(txt):
  #no_punct = "".join([c for c in txt if c not in string.punctuation])
  #tokens = re.split('\W+',txt)
  txt = " ".join([wn.lemmatize(word) for word in txt if word not in stop])
  return txt
```

```python
df['clean_review_sentence'] = df['review'].apply(lambda x: clean_text_sentence(x))
```
